```python
#first we need to import the data and import the correct libraries
import pandas as pa
import matplotlib.pyplot as plt
import seaborn as sns
df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/worldCupPlayer.csv')
df
```

Since I pulled the data, let me pick a random nationality. Sweden! Here is their flag.

sweden.png

```python
#Question1b Lets pull everything from Sweden!
df2=df.loc[df['nationality']=='Sweden']
df2
```

```python
#Question 1c the code below shows the goals on a histogram

df.Goals.plot.hist(bins= 30, title="Goals")
```

```python
#Question 1d the code below shows the goals and assists all of the same axis and the .9 shows the thickness of the color

df.groupby('Goals').Assists.plot.hist(alpha = .9)
```

```python
#Question 1e
df2 = df.groupby('nationality').Goals.agg(count = 'count') #This section will grab all of the nationalities from the dataframe
df2 = df2.reset_index()

df2
```

```python
#Now in this part, I'll create a barchart of the goals scored by nationality
ax = sns.barplot(data = df2,x = 'nationality', y = 'count')
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
ax.set_title('Goals made from Nationality')

plt.show()
```

```python
#Question 1f lets create a boxplot of Yel 
df.boxplot(column='Yel', by='nationality', rot=90)
plt.show()
```

```python
#Question 1g creating a scatterplot 
ax = sns.scatterplot(data = df, 
                x = 'Goals', 
                y = 'Assists',
                hue = 'Goals',
                size = 'nationality',
                legend = 'brief')
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
```

```python
#Question 1i #this is the heatmap of the entire data frame
sns.heatmap(df.corr(), annot=True, linewidths=2.5,vmin = -0.5) #I wanted my minimum to be -0.5
```
