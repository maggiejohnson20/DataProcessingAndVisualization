Let's start the Exam 2 by listing the libraries that we need to use and pull the dataframe.

```python
import pandas as pa
import numpy as np
import re

df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')

df.head()
```

```python
df.dtypes
```

For part b, we need to remove the columns that repeats the indexes and is 'unnamed' 

In the code below, drop means to delete

```python
df2 = df.drop('Unnamed: 0', axis =1) 

df2
```

For part c, we need to clean up the column names. Let's get rid of the unit declaration!

In the code below, lamba helps us remove the unit declaration.

```python
df2 = df.rename(columns = lambda x: re.sub(r'((.%))' ,'' ,x))

df2.columns
```

Since we have that figured out, we need to start on part d, which is converting activity_day column into a dataframe format.

The code below first gave us the data type and we can use that to covert the activity!

```python
df2.activity_day.dtypes
```

```python
df2.activity_day = pa.to_datetime(df2.activity_day)
df2.activity_day.dtypes
```

Now we need to focus on part e of the exam! In this part we need to impute the total_steps, get rid of all of the NaN's by replacing it with appropriate number of steps. Then covert it like we did in the steps above.

To do that, we use the fillna a function to replace the NaN's with the minimum number of steps.

```python
df2.total_steps.fillna(value=df.total_steps.min(), inplace = True)

df2.head()
```

For part F, we need to convert the nonempty 'avg-pace' into a float

```python
df2.avg_pace = df2.avg_pace.fillna(0)

df2.info
```

For part G, let's find the mean, median, count, and standard deviation of the workout_time.

```python
df2.groupby('workout_type').calories.agg(['mean','median','count', 'std'])
```

For part H, let's indicate whether the activity 30% or more aerobic

```python
df2 = df2.loc[df2['max_cadence'] == df2['max_cadence'].max()]
print('day_of_week = Tuesday' )
print('workout_type = ' + df2.workout_type.iloc[0])
print('maximum_cadence = ' + str(df2['max_cadence'].max()))

```
