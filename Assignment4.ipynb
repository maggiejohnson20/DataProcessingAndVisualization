For this assignment, one part is to find the average time of the trips that are based off of the starting points. 

```python
%%bigquery --project dataprocessingvisualization

SELECT AVG(duration_minutes) as avg_time, start_station_name
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`

GROUP BY start_station_name
Order by avg_time

```

Once we've figured that out, it's time to figure out the next part of the assignment. We need to figure out how many number of trips were taken at each start stations. 

```python
%%bigquery --project dataprocessingvisualization

SELECT COUNT(start_station_name) as No_Trip, start_station_name
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`

GROUP BY start_station_name
Order by No_Trip
```
