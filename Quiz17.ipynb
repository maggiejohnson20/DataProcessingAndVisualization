Let's begin by grabbing the data frame.

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/iris.csv')
df
```

Lets try to see if we can find the mean, median, count and sd if we group by class.

```python
df.groupby('Class').SepalLength.agg(['mean','median','count', 'std'])
```
