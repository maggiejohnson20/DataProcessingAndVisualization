```python
import pandas as pa
import matplotlib.pyplot as plt
import seaborn as sns
df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/bank.csv')
df
```

```python
ax = sns.scatterplot(data = df, 
                x = 'marital', 
                y = 'job',
                size = 'balance',
                hue = 'duration',
                legend = 'brief')
sns.move_legend(ax, "upper left", bbox_to_anchor=(1.1,1))
```

```python
sns.set(rc={"figure.figsize":(20, 10)})
sns.heatmap(df.corr(), annot=True, linewidths=1.5)
```

```python
sns.boxplot(data = df, y = 'balance', x = 'job')
```
