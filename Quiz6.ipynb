Now it's time to actually work on the project. In this quiz, we need to see if the average time is longer on parkland or on sidewalk.

```python
%%bigquery --project dataprocessingvisualization

WITH tbl as(
SELECT name, property_type
FROM `bigquery-public-data.austin_bikeshare.bikeshare_stations`
)

SELECT tbl.property_type as Starting_Station_Types, 
    AVG (trips.duration_minutes) as Average_Minutes,
    COUNT(*) as Number_of_Trips,
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips` as trips
  JOIN tbl
  ON trips.start_station_name = tbl.name

GROUP BY tbl.property_type
ORDER by Average_Minutes
```
