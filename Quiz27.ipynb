```python
import pandas as pa
import matplotlib.pyplot as plt
import seaborn as sns

df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')
df.head()
```

```python
df1 = df.groupby('workout_type').workout_type.agg(count = 'count')
df1
```

```python
plt.pie(x=df.groupby('workout_type').workout_type.agg('count'),labels = df.groupby('workout_type').workout_type.agg('count').index, autopct='%.0f%%' )

plt.show()
```

```python
#Create the bar graph by gathering the data first
x=df.groupby('workout_type').workout_type.agg('count')
labels = df.groupby('workout_type').workout_type.agg('count').index

#Now we should resize everything to make it better to look at
fig, ax = plt.subplots(figsize=(15, 30)) 
ax.pie(x, labels=labels)
plt.legend(labels,loc=2)
ax.set_title("Types of Workouts")
```

One of the things that a Mosaic graph can show that a bar or pie chart can't is the different variables in the single object. The mosaic graph can use the length to represent one thing and the width to represent another thing. That's why the graph looks so odd. The bar and pie charts cannot do that, so that's why they look the same width-wise.
